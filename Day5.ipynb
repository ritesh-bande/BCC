```python
#function
def add():
  n1=int(input("Enter the value of n1:"))
  n2=int(input("Enter the value of n2:"))
  sum=n1+n2
  mul=n1*n2
  sub=n1-n2
  div=n1/n2
  return sum,mul,sub,div
result=add()
print(result)
```

```python
# #4 types of arguement can be passed
# 1.positional arguement
# 2.keyword
# 3.default
# 4.variable length /variable number of arguement
```

```python
#1.positional arguement
def personalInfo(fname,lname):
  print("First Name:",fname)
  print("Last Name:",lname)
personalInfo("Ritesh",'Bande')
```

```python
# 2.keyword arguement
#1.positional arguement
def personalInfo(fname,lname):
  print("First Name:",fname)
  print("Last Name:",lname)
fname="Ritesh"
lname="Bande"
personalInfo(fname,lname)
```

```python
#default arguement
def cityName(city="Nagpur"):
  print(city)
cityName("Mumbai")
cityName("Delhi")
cityName()
```

```python
#variable length
def StudentNames(*name):
  print(name)
  print(type(name))
StudentNames("Pk",'Sujal','kunal','Albash')
```

```python
mylist=[5,2,9,7,5,6]
def searchElement(target):
  for i in range(len(mylist)):
    if target==mylist[i]:
      print("Element found at index-",i)
searchElement(7)
```

```python
myList=[5,2,9,7,5,6]
def searchElement(target):
  for i in range(len(myList)):
    if target==mylist[i]:
      return i
  return -1

result=searchElement(10)
if result!=1:
  print("Element found",result)
else:
  print("Element not found.")
```
