```python
name = "racecar"

string_length = len(name)
is_palindrome = True

for i in range(string_length // 2):
    if name[i] != name[string_length - 1 - i]:
        is_palindrome = False
        break

if is_palindrome:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```
